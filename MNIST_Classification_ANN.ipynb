```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
x_train
```

```python
y_train
```

```python
import matplotlib.pyplot as plt

plt.imshow(x_train[0])
```

```python
model=Sequential()
#Input Dimension is 28x28
# using flatten for 2d data which is image it will flatten the 2d data into 1d or into nueral network form
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer='Adam')
```

```python
model.fit(x_train, y_train, epochs=50, validation_split=0.2)
```

```python
prob=model.predict(x_test)
```

```python
y_pred=prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

We can also imporve our model by adding extra layer/nodes

```python
model2=Sequential()

model2.add(Flatten(input_shape=(28,28)))
model2.add(Dense(128, activation='relu'))
model2.add(Dense(32, activation='relu'))
model2.add(Dense(10, activation='softmax'))

```

```python
model2.summary()
```

```python
model2.compile(loss='sparse_categorical_crossentropy', optimizer='Adam')
```

```python
model_logs=model2.fit(x_train, y_train, epochs=25, validation_split=0.2)
```

```python
y_prob=model2.predict(x_test)
```

```python
ypred=y_prob.argmax(axis=1)
```

```python
accuracy_score(y_test, ypred)
```
